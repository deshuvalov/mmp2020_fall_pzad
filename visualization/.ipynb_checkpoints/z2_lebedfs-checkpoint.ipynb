```python
from itertools import *
from functools import *

import numpy as np
import pandas as pd

import scipy.stats as stats
import pingouin as pg

import matplotlib.pyplot as plt
import seaborn as sns

IMG_FORMAT = 'png'
```

```python
## Utils
from functools import partial as p
from functools import reduce as r
def c(*fs): return p(r, (lambda y,f: f(y)), fs)

def clean_names(df): return df.rename(columns={col:col.strip().replace(' ', '_') for col in df.columns})
def rm_duplicates(df): return df[~df.duplicated()]
def m(f, *args, **kwargs): return (lambda x: f(x, *args, **kwargs))
def nth(n): return (lambda v: v[n])
fst = nth(0)
def get(attr): return (lambda o: getattr(o, attr))
def mkseries(v): return pd.Series(v, index=v)
def to_year(t): return t.year + (t.month - 1)/12
def nth_largest(df, *, key=None, col=None, n=None): return df.groupby(key)[col].nlargest(n).reset_index().groupby(key)[col].min()
def chain_dict(*ds): return {k:v for k,v in chain(*list(map(dict.items, ds)))}


## Plotting
def plot1d(s, ax=None, kind='bar', title=None, xlabel=None, ylabel=None, **kwargs):
    subplot = ax is not None
    
    if not subplot:
        fig, ax = plt.subplots(figsize=(10*1.5, 10))
    
    if kind == 'bar':
        s.plot.bar(ax=ax, **chain_dict(barstyle, kwargs))
    elif kind == 'hist':
        s.plot.hist(ax=ax, **chain_dict(barstyle, kwargs))
    else:
        raise Exception('no such kind: {}'.format(kind))
        
    ax.set_xlabel(xlabel)
    ax.set_ylabel(ylabel)
    
    if not subplot:
        fig.suptitle(title)
        fig.tight_layout()
        fig.savefig('pres/img/{}.{}'.format(title.replace('\n', ' '), IMG_FORMAT))
```

```python
plt.style.use('seaborn')
barstyle = dict(edgecolor='black', linewidth=0.5)
```

```python
raw_likes = c(pd.read_csv, clean_names)('data/2018_ratings.csv')
raw_users = c(pd.read_csv, clean_names)('data/2018-personality-data.csv')
```

```python
user_features = ['userid', 'openness', 'agreeableness', 'emotional_stability', 'conscientiousness', 'extraversion']

likes = rm_duplicates(raw_likes).rename(columns=dict(useri='userid'))
likes = likes.assign(
    tstamp=likes.tstamp.map(c(str.split, fst, pd.to_datetime))
)
likes = likes.assign(
    year=likes.tstamp.map(get('year'))
)
```

```python
pd.Series(range(15)).nlargest(3)
```

```python
users = rm_duplicates(raw_users[user_features])
users = users.assign(
    first_year=users.userid.map(likes.groupby('userid').year.min()),
    first_tstmp=users.userid.map(likes.groupby('userid').tstamp.min()).map(to_year),
    
    last_year=users.userid.map(likes.groupby('userid').year.max()),
    last_tstmp=users.userid.map(likes.groupby('userid').tstamp.max()).map(to_year),
    
    prelast_tstmp=users.userid.map(nth_largest(likes, key='userid', col='tstamp', n=12+1)).map(to_year),    
    movie_cnt=users.userid.map(likes.groupby('userid').size())
)
users = users.assign(
    lifetime=(users.last_tstmp-users.first_tstmp),
    prelifetime=(users.prelast_tstmp-users.first_tstmp)
).set_index('userid')
```

```python
movies = pd.DataFrame(
    dict(
        first_year=likes.groupby('movie_id').year.min(),
        first_tstmp=likes.groupby('movie_id').tstamp.min().map(to_year),
        
        last_year=likes.groupby('movie_id').year.max(),
        last_tstmp=likes.groupby('movie_id').tstamp.max().map(to_year),
        
        prelast_tstmp=nth_largest(likes, key='movie_id', col='tstamp', n=1+1).map(to_year),
        user_cnt=likes.groupby('movie_id').size()
    )
)
movies = movies.assign(
    lifetime=(movies.last_tstmp-movies.first_tstmp),
    prelifetime=(movies.prelast_tstmp-movies.first_tstmp)
)
```

```python
agg_likes = likes.assign(
    user_year=likes.userid.map(users.first_year),
    movie_year=likes.movie_id.map(movies.first_year)
)
```

```python
assert users.index.nunique() == len(users)
assert movies.index.nunique() == len(movies)
assert likes.groupby(['userid', 'movie_id']).size().max() == 1
```

```python
fs = ['openness', 'agreeableness', 'emotional_stability', 'conscientiousness', 'extraversion']
fig, ax = plt.subplots(figsize=(15, 10))

users[fs + ['first_year']].groupby('first_year').mean().plot.box(ax=ax, vert=False)
ax.set_xlim(0, 7)

ax.set_xlabel('значение метрики')
ax.set_ylabel('название метрики')
plt.suptitle('Распределения средних за год\nзначений метрик личности')
plt.tight_layout()
plt.savefig('pres/img/Распределения средних за год значений метрик личности.{}'.format(IMG_FORMAT))
```

```python
def plot_metric(metrics, *, ax=None, kind='bar', **kwargs):
    for i, f in enumerate(fs):
        if kind == 'bar':
            (metrics.groupby(f).size() / len(metrics)).plot.barh(ax=ax[i], **chain_dict(barstyle, kwargs))
        elif kind == 'hist':
            ax[i].hist(metrics[f], density=True, orientation='horizontal', **chain_dict(barstyle, kwargs))
        ax[i].set_ylabel('' if i else 'значение метрики')
        ax[i].set_xlabel('доля значений')
        ax[i].set_title(f)        
```

```python
fig, ax = plt.subplots(1, 5, figsize=(20, 10))
plot_metric(users, ax=ax)
fig.suptitle('Распределения значений метрик личности')
fig.tight_layout()
fig.savefig('pres/img/Распределения значений метрик личности.{}'.format(IMG_FORMAT))
```

```python
fig, ax = plt.subplots(2, 5, figsize=(20, 10))
cbar_ax = fig.add_axes([1.01, 0.25, .01, 0.5])
for i, (f1, f2) in enumerate(combinations(fs, 2)):
    sns.histplot(users, binwidth=0.5, discrete=True, x=f1, y=f2, ax=ax[i//5, i%5], 
                 cbar=not bool(i), cbar_ax=cbar_ax)
    ax[i//5, i%5].set_xticks(range(1, 7+1))
    ax[i//5, i%5].set_yticks(range(1, 7+1))
    ax[i//5, i%5].grid(0)

plt.suptitle('Таблица сопряженности для всех пар метрик личности')
plt.tight_layout()
plt.savefig('pres/img/Таблица сопряженности для всех пар метрик личности.{}'.format(IMG_FORMAT))
```
